### Create a Lambda Function to give marks to students based on Time of Submission of Responses (5, 4, 3, 2, 1, 0, etc) and then create a new column based on that

```python
# giving mark for punctuality
def time_mark(time):
    if int(time[13:15]) < 9:
        return 1
    elif int(time[13:15]) < 11:
        return 2
    elif int(time[13:15]) < 13:
        return 3
    return 4
```

```python
df['punc_mark'] = df['Timestamp'].apply(lambda time : time_mark(time))
```

### Create a new column Programming, with the values "Low" (1), "Medium" (2, 3), "Good" (4, 5), based on the ProgrammingLevel Column.

```python
# level of programming
def prog_level(num):
    if num == 1:
        return 'Low'
    elif 1 < num < 4:
        return 'Medium'
    return 'Good'

df['Program_level'] = df['Programming Level ?(for 5)'].apply(lambda num : prog_level(num))
df.head(15)
```
